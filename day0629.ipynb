## 행과 열 다루기
- Sold(판매량) 컬럼 (=피쳐=feature)을 추가


```python
lemonade['Sold'] = 0  #lemonade.Sold으로도 가능하지만, 메서드인지 칼럼인지 헷갈리는 경우가 있음
print(lemonade.head(3))
```

```python
lemonade['Sold'] = lemonade['Lemon'] + lemonade['Orange']
print(lemonade.head(3))
```

- Revenue(매출) = 단가 x 판매량 

```python
lemonade['Revenue'] = 0
print(lemonade.head(3))
```

```python
lemonade['Revenue'] = lemonade['Price'] * lemonade['Sold']
print(lemonade.head(3))

```

- drop 함수 사용해서 열 제거

```python
# 컬럼 제거 
col_drop = lemonade.drop('Sold', axis=1)
print(col_drop.head()) 
```

```python
# 행 제거
row_drop = lemonade.drop(2, axis = 0)
print(row_drop.head()) 
```

```python
# 범위 제거
row_drop = lemonade.drop([0,1,2], axis = 0)
print(row_drop.head()) 
```

## iloc vs loc
-문법상의 차이

```python
lemonade.head()
```

- iloc
 + 숫자만 들어간다
 + 필터링시 사용이 용이하지 않다

```python
print(lemonade.iloc[0:3,0:2])
```

- loc: 숫자 (X)
 + 라벨= 글자(숫자,문자 동시)
 + 사용 추천

```python
print(lemonade.loc[0:2,['Date','Location']])
```
